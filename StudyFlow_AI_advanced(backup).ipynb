```python
# ============================================================
# CELL 5 вҖ” Launch StudyFlow-AI  (Cloudflare tunnel вҖ” free)
# ============================================================

import subprocess, time, re

# 1. Download cloudflared binary
print("в¬ҮпёҸ  Downloading cloudflaredвҖҰ")
import subprocess
subprocess.run(["wget", "-q",
    "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64",
    "-O", "cloudflared"], check=True)
subprocess.run(["chmod", "+x", "cloudflared"], check=True)
print("вң… cloudflared ready")

# 2. Start Streamlit
print("\nрҹҡҖ Starting StudyFlow-AI (Streamlit)вҖҰ")
streamlit_proc = subprocess.Popen(
    [
        "streamlit", "run", "main.py",
        "--server.port",                 "8501",
        "--server.headless",             "true",
        "--server.enableCORS",           "false",
        "--server.enableXsrfProtection", "false",
        "--server.maxUploadSize",        "200",
    ],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL,
)
time.sleep(4)
print("вң… Streamlit running on port 8501")

# 3. Start Cloudflare tunnel
print("\nрҹҢҗ Opening Cloudflare tunnelвҖҰ")
tunnel_proc = subprocess.Popen(
    ["./cloudflared", "tunnel", "--url", "http://localhost:8501"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    universal_newlines=True,
)

print("вҸі Waiting for tunnel URL (this takes ~10 s)вҖҰ\n")
for line in tunnel_proc.stdout:
    match = re.search(r'https://[a-z0-9\-]+\.trycloudflare\.com', line)
    if match:
        url = match.group(0)
        print("=" * 55)
        print("  вң…  StudyFlow-AI is live at:")
        print(f"      {url}")
        print("=" * 55)
        print("\nOpen that URL in any browser вҖ” no password needed.")
        break

# Keep tunnel alive (interrupt this cell to stop)
tunnel_proc.wait()
```
